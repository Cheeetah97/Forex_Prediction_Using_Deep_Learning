```python
import pandas as pd
import numpy as np
from sklearn.metrics import precision_score,accuracy_score,mean_squared_error
import matplotlib.pyplot as plt

p1 = "M:/Dissertation/Price_Prediction/Results/"
p2 = "M:/Dissertation/Return_Prediction/Deep_Learning/Results/"
p3 = "M:/Dissertation/Return_Prediction/Machine_Learning/Results/"
p4 = "M:/Dissertation/Return_Prediction/Transformer/Results/"
```

### **Results - Return Prediction Using LSTM**

```python
results_ret_dl = pd.read_csv(p2+"Test_Results.csv")

targets = pd.melt(results_ret_dl,id_vars=['Date','Portion'],value_vars=['EUR/USD_T','EUR/GBP_T','GBP/USD_T','XAU/USD_T'],value_name="Return",var_name="Currency")
targets["Currency"] = targets["Currency"].str.split("_",expand=True)[0]

preds = pd.melt(results_ret_dl,id_vars=['Date','Portion'],value_vars=['EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P'],value_name="Predicted_Return_DL",var_name="Currency")
preds["Currency"] = preds["Currency"].str.split("_",expand=True)[0]

prices = pd.melt(results_ret_dl,id_vars=['Date','Portion'],value_vars=['EUR/USD','EUR/GBP','GBP/USD','XAU/USD'],value_name="Price",var_name="Currency")
prices["Currency"] = prices["Currency"].str.split("_",expand=True)[0]

preds_p = pd.melt(results_ret_dl,id_vars=['Date','Portion'],value_vars=['EUR/USD_PP','EUR/GBP_PP','GBP/USD_PP','XAU/USD_PP'],value_name="Predicted_Price_DL",var_name="Currency")
preds_p["Currency"] = preds_p["Currency"].str.split("_",expand=True)[0]

results_ret_dl = targets.merge(prices,how="left",on=["Date","Portion","Currency"])
results_ret_dl = results_ret_dl.merge(preds,how="left",on=["Date","Portion","Currency"])
results_ret_dl = results_ret_dl.merge(preds_p,how="left",on=["Date","Portion","Currency"])
results_ret_dl = results_ret_dl.dropna()
results_ret_dl
```

### **Results - Price Prediction Using LSTM**

```python
results_price = pd.read_csv(p1+"Test_Results.csv")
results_price = results_price[['Date','Portion','EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P','EUR/USD_PR','EUR/GBP_PR','GBP/USD_PR','XAU/USD_PR']]
results_price["Date"] = results_price.groupby(by=["Portion"],as_index=False)["Date"].shift(-1)

preds_p = pd.melt(results_price,id_vars=['Date','Portion'],value_vars=['EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P'],value_name="Predicted_Price_DLP",var_name="Currency")
preds_p["Currency"] = preds_p["Currency"].str.split("_",expand=True)[0]

returns_p = pd.melt(results_price,id_vars=['Date','Portion'],value_vars=['EUR/USD_PR','EUR/GBP_PR','GBP/USD_PR','XAU/USD_PR'],value_name="Predicted_Return_DLP",var_name="Currency")
returns_p["Currency"] = returns_p["Currency"].str.split("_",expand=True)[0]

results_price = returns_p.merge(preds_p,how="left",on=["Date","Portion","Currency"])
results_price = results_price.dropna()
results_price
```

### **Return Prediction Using LGBM**

```python
results_ret_ml = pd.read_csv(p3+"Test_Results.csv")
results_ret_ml = results_ret_ml[['Date','Portion','EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P','EUR/USD_PP','EUR/GBP_PP','GBP/USD_PP','XAU/USD_PP']]

preds = pd.melt(results_ret_ml,id_vars=['Date','Portion'],value_vars=['EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P'],value_name="Predicted_Return_ML",var_name="Currency")
preds["Currency"] = preds["Currency"].str.split("_",expand=True)[0]

preds_p = pd.melt(results_ret_ml,id_vars=['Date','Portion'],value_vars=['EUR/USD_PP','EUR/GBP_PP','GBP/USD_PP','XAU/USD_PP'],value_name="Predicted_Price_ML",var_name="Currency")
preds_p["Currency"] = preds_p["Currency"].str.split("_",expand=True)[0]

results_ret_ml = preds.merge(preds_p,how="left",on=["Date","Portion","Currency"])
results_ret_ml = results_ret_ml.dropna()
results_ret_ml
```

### **Return Prediction Using Informer**

```python
results_ret_inf = pd.read_csv(p4+"Test_Results.csv")
results_ret_inf = results_ret_inf.rename(columns={"Test_Portion":"Portion"})

preds = pd.melt(results_ret_inf,id_vars=['Date','Portion'],value_vars=['EUR/USD_P','EUR/GBP_P','GBP/USD_P','XAU/USD_P'],value_name="Predicted_Return_INF",var_name="Currency")
preds["Currency"] = preds["Currency"].str.split("_",expand=True)[0]

preds_p = pd.melt(results_ret_inf,id_vars=['Date','Portion'],value_vars=['EUR/USD_PP','EUR/GBP_PP','GBP/USD_PP','XAU/USD_PP'],value_name="Predicted_Price_INF",var_name="Currency")
preds_p["Currency"] = preds_p["Currency"].str.split("_",expand=True)[0]

results_ret_inf = preds.merge(preds_p,how="left",on=["Date","Portion","Currency"])
results_ret_inf = results_ret_inf.dropna()
results_ret_inf
```

### **Combining All Results**

```python
all_results = results_ret_dl.merge(results_ret_ml,how="left",on=["Date","Portion","Currency"])
all_results = all_results.merge(results_price,how="left",on=["Date","Portion","Currency"])
all_results = all_results.merge(results_ret_inf,how="left",on=["Date","Portion","Currency"])
all_results = all_results.dropna().sort_values(by=["Date"]).reset_index(drop=True)
all_results.to_csv('All_Results.csv',index=False)
```

### **Comparing Mean Squared Errors and Accuracy Scores**

```python
metrics = pd.DataFrame()
for c in list(all_results.Currency.unique()):
    temp = all_results[all_results.Currency==c]
    
    for p in list(temp.Portion.unique()):
        temp2 = temp[temp.Portion==p]
        temp2 = temp2.sort_values(by=["Date"])
        temp2 = temp2.reset_index(drop=True)
        orig_cols = temp2.drop(["Currency","Portion"],axis=1).columns

        # Mean Squared Errors
        temp2["MSE(LSTM_Return)"] = mean_squared_error(temp2["Return"],temp2["Predicted_Return_DL"])
        temp2["MSE(LGBM_Return)"] = mean_squared_error(temp2["Return"],temp2["Predicted_Return_ML"])
        temp2["MSE(LSTM_Price)"] = mean_squared_error(temp2["Return"],temp2["Predicted_Return_DLP"])
        temp2["MSE(INFORMER_Return)"] = mean_squared_error(temp2["Return"],temp2["Predicted_Return_INF"])
        temp2["MSE(MEAN_Return)"] = mean_squared_error(temp2["Return"],pd.Series([temp2["Return"].mean() for i in range(len(temp2))]))
        
        # Accuracy Scores
        temp2["Price_1"] = temp2["Price"].shift(-1)
        temp2["Predicted_Price_DL_1"] = temp2["Predicted_Price_DL"].shift(-1)
        temp2["Predicted_Price_ML_1"] = temp2["Predicted_Price_ML"].shift(-1)
        temp2["Predicted_Price_DLP_1"] = temp2["Predicted_Price_DLP"].shift(-1)
        temp2["Predicted_Price_INF_1"] = temp2["Predicted_Price_INF"].shift(-1)

        temp2["Price_Inc"] = 0
        temp2.loc[temp2["Price_1"] > temp2["Price"],"Price_Inc"] = 1

        temp2["Price_Inc_DL"] = 0
        temp2.loc[temp2["Predicted_Price_DL_1"] > temp2["Predicted_Price_DL"],"Price_Inc_DL"] = 1

        temp2["Price_Inc_ML"] = 0
        temp2.loc[temp2["Predicted_Price_ML_1"] > temp2["Predicted_Price_ML"],"Price_Inc_ML"] = 1

        temp2["Price_Inc_DLP"] = 0
        temp2.loc[temp2["Predicted_Price_DLP_1"] > temp2["Predicted_Price_DLP"],"Price_Inc_DLP"] = 1

        temp2["Price_Inc_INF"] = 0
        temp2.loc[temp2["Predicted_Price_INF_1"] > temp2["Predicted_Price_INF"],"Price_Inc_INF"] = 1

        temp2["ACC_SCRE(LSTM_Return)"] = accuracy_score(temp2["Price_Inc"],temp2["Price_Inc_DL"])
        temp2["ACC_SCRE(LGBM_Return)"] = accuracy_score(temp2["Price_Inc"],temp2["Price_Inc_ML"])
        temp2["ACC_SCRE(LSTM_Price)"] = accuracy_score(temp2["Price_Inc"],temp2["Price_Inc_DLP"])
        temp2["ACC_SCRE(INFORMER_Return)"] = accuracy_score(temp2["Price_Inc"],temp2["Price_Inc_INF"])
        
        temp2 = temp2.drop(list(orig_cols)+['Price_1','Predicted_Price_DL_1','Predicted_Price_ML_1','Predicted_Price_DLP_1','Predicted_Price_INF_1','Price_Inc',
                                            'Price_Inc_DL','Price_Inc_ML','Price_Inc_DLP','Price_Inc_INF'],axis=1)
        temp2 = temp2.drop_duplicates()
        metrics = pd.concat([metrics,temp2])
    
mses = metrics.groupby(by=["Currency"],as_index=False)[['MSE(LSTM_Price)','MSE(LSTM_Return)','MSE(LGBM_Return)','MSE(INFORMER_Return)','MSE(MEAN_Return)']].mean()
accuracies = metrics.groupby(by=["Currency"],as_index=False)[['ACC_SCRE(LSTM_Price)','ACC_SCRE(LSTM_Return)','ACC_SCRE(LGBM_Return)','ACC_SCRE(INFORMER_Return)']].mean()

mses.to_csv('MeanSquaredErrors.csv',index=False)
accuracies.to_csv('AccuracyScores.csv',index=False)
```

### **Evaluating the Results in a Trading Scenario**

```python
inv_amount = 20000

trading_hist = pd.DataFrame()
for c in list(all_results.Currency.unique()):
    temp = all_results[all_results.Currency==c]
    
    for p in list(temp.Portion.unique()):
        temp2 = temp[temp.Portion==p]
        temp2 = temp2.sort_values(by=["Date"])
        temp2 = temp2.reset_index(drop=True)

        temp2 = temp2[["Date","Currency","Portion","Price","Predicted_Price_ML"]]
        
        # Price Increment and Decrement Indicators
        temp2["Price_Inc"] = 0
        temp2["Price_1"] = temp2["Price"].shift(-1)
        temp2["Price_Change"] = np.abs(temp2["Price_1"] - temp2["Price"])
        temp2.loc[temp2["Price_1"] > temp2["Price"],"Price_Inc"] = 1

        temp2["Price_Inc_ML"] = 0
        temp2["Predicted_Price_ML_1"] = temp2["Predicted_Price_ML"].shift(-1)
        temp2.loc[temp2["Predicted_Price_ML_1"] > temp2["Predicted_Price_ML"],"Price_Inc_ML"] = 1
        
        # Calculating Profit
        temp2["Buy"] = 0
        temp2.loc[temp2["Price_Inc_ML"]==1,"Buy"] = temp2["Price_Inc"] * inv_amount * temp2["Price_Change"]

        temp2["Sell"] = 0
        temp2.loc[temp2["Price_Inc_ML"]==0,"Sell"] = (1-temp2["Price_Inc"]) * inv_amount * temp2["Price_Change"]

        temp2["Profit"] = temp2["Buy"] + temp2["Sell"] 

        # Calculating Loss
        temp2["Buy"] = 0
        temp2.loc[temp2["Price_Inc_ML"]==1,"Buy"] = (1-temp2["Price_Inc"]) * inv_amount * temp2["Price_Change"]

        temp2["Sell"] = 0
        temp2.loc[temp2["Price_Inc_ML"]==0,"Sell"] = temp2["Price_Inc"] * inv_amount * temp2["Price_Change"]

        temp2["Loss"] = -(temp2["Buy"] + temp2["Sell"]) 

        # Adjusting Transaction Costs
        temp2["Profit"] = temp2["Profit"] - 1
        temp2["Loss"] = temp2["Loss"] - 1

        # Net Profit
        temp2["Net_Profit"] = temp2["Profit"] + temp2["Loss"]
        temp2["Net_Profit"] = temp2["Net_Profit"].cumsum()

        trading_hist = pd.concat([trading_hist,temp2[["Date","Currency","Portion","Net_Profit"]]])

trading_hist.to_csv("Trading_Hist_WithCost.csv",index=False)
```
